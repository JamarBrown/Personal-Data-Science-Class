```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics._plot import confusion_matrix
```

```python
#import dataset

data = datasets.load_iris()

x = data.data[:,2:4]
y = data.target
```

```python
svm_model = svm.SVC(kernel="linear", C=0.1, gamma='auto')
svm_model.fit(x,y)
```

```python
#Creating the mesh or grid to plot the samples and hyperplane

x_min, x_max = x[:,0]. min()-0.5, x[:,0].max()-0.5
y_min, y_max = x[:,1]. min()-0.5, x[:,1].max()-0.5

h= (x_max/x_min)/100
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),(y_min, y_max, h))
```

```python
plt.subplot(1,1,1)
classified_labels = svm_model.predict(np.c_[xx.ravel(),yy.ravel()])
classified_labels = classified_labels.reshape(xx.shape)
plt.contour(xx,yy, classified_labels, cmp=plt.cm.Paired, alpha=0.8)
plt.scatter(x[:,0], x[:,1],c=y, cmap=plt.cm.Paired)
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.xlim(xx.min(),xx.max())
plt.title('SVC with Linear Kernel')
plt.show()
```

```python
#Divide the data into training and testing

X=data.data
Y=data.target

x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.2, random_state=33)
svm_model2 = svm.SVC(kernel='linear', C=1, gamma='auto')

svm_model2.fit(x_train, y_train)
```

```python
classification_results = svm_model2.predict(x_test)

print(metrics.classification_report(classification_results,y_test))
```
