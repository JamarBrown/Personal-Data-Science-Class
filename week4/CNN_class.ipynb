```python
import matplotlib.pyplot as plt
import numpy as np
from keras.datasets import fashion_mnist
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

```python
(x_train,y_train), (x_test, y_test) = fashion_mnist.load_data()


classes = np.unique(y_train)
nClasses = len(classes)
print("This dataset has {0:d} classes".format(nClasses))
 
y_name = {0:"T-Shirt",1:"Trousers", 2:"Pullover", 3:"Dress",4:"Coat", 5:"Sandals", 6:"Shirt", 7:"Sneaker", 8:"Bag", 9:"Ankle Boot" }

print("Image Size: ", x_train[0].shape)
```

```python
_, fig = plt.subplots(1,3,figsize=(10,10))

for ax, image, label in zip(fig, x_train, y_train):
    ax.set_axis_off()
    ax.imshow(image, cmap=plt.cm.Oranges)
    ax.set_title( "Training: %s" %y_name[label])
```
